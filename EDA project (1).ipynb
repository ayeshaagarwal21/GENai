```python
#Use deliveries data frame and find out year by year how many number of runs have we scored in each season, Now get the total no of runs score in each season by each team,now plot line chart
merged_df=pd.merge(df_deliveries_data,match_data[['id','season']],left_on='id',right_on='id') #merge is used to merge the files 
season_team_runs=merged_df.groupby(['season','batting_team'])['total_runs'].sum().reset_index() #reset_index is used for saving the calculated result into a clean dataframe
pivot_df=season_team_runs.pivot(index='season',columns='batting_team',values='total_runs') #converting data into matrix form
plt.figure(figsize=(15,8))
for team in pivot_df.columns:
    plt.plot(pivot_df.index,pivot_df[team],marker='o',label=team) #marker is used for pointers
plt.title('Total Runs Scored by Each Team in Each IPL Season')
plt.xlabel('Season')
plt.ylabel('Total Runs')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Get the overall runs score by virat kohli in once in two in threes in fours and in sixes
df_deliveries_data.columns = df_deliveries_data.columns.str.strip() #remove white spaces from the strings
kohli_df = df_deliveries_data[df_deliveries_data['batsman'] == 'V Kohli']
run_counts = {  #counts the no of runs made by virat
    'Ones': kohli_df[kohli_df['batsman_runs'] == 1].shape[0],
    'Twos': kohli_df[kohli_df['batsman_runs'] == 2].shape[0],
    'Threes': kohli_df[kohli_df['batsman_runs'] == 3].shape[0],
    'Fours': kohli_df[kohli_df['batsman_runs'] == 4].shape[0],
    'Sixes': kohli_df[kohli_df['batsman_runs'] == 6].shape[0],
}
pd.Series(run_counts).plot(kind='bar', color='orange', edgecolor='black', figsize=(8, 6))
plt.title('Virat Kohli - Run Type Distribution in IPL (2008–2020)')
plt.xlabel('Run Type')
plt.ylabel('Number of Times')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
#Get the top five wicket keepers based on lbw
import matplotlib.pyplot as plt
wicket_data=df_deliveries_data[(df_deliveries_data["dismissal_kind"]=='lbw')& (df_deliveries_data['batsman'].notna())]
f=wicket_data.head(5)
print("Top 5 wicketkeepers based on LBW dismissals:")
print(wicket_data)
plt.figure(figsize=(10, 6))
f.plot(kind='bar', color='yellow', edgecolor='black')
plt.title('Top 5 Wicketkeepers Based on LBW Dismissals in IPL')
plt.xlabel('Wicketkeeper (Fielder)')
plt.ylabel('Number of LBW Dismissals')
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
#Find out each team how many number of matches won in 2016 and compare it to 2017,plot graph
match_data.columns = match_data.columns.str.strip()
match_data['date'] = pd.to_datetime(match_data['date'])
match_data['season'] = match_data['date'].dt.year
match_wins = match_data[match_data['season'].isin([2016, 2017])] #put data of 2016 and 17 in match wins
wins_by_team_year = match_wins.groupby(['season', 'winner']).size().unstack(fill_value=0)
wins_by_team_year = wins_by_team_year.transpose() #to transpose x and y axis
wins_by_team_year.plot(kind='bar', figsize=(12, 6), color=['orange', 'black'])
plt.title('Matches Won by Each Team in 2016 vs 2017')
plt.xlabel('Team')
plt.ylabel('Matches Won')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.6) #alpha is used for width
plt.tight_layout()
plt.show()
```
